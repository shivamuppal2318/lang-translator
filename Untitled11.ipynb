```python
def translate_text(text, model, tokenizer):
    # Tokenize the input text
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)

    # Generate translation
    translated = model.generate(**inputs)

    # Decode the translated text
    translated_text = tokenizer.decode(translated[0], skip_special_tokens=True)
    return translated_text
```

```python
import torch
from transformers import MarianMTModel, MarianTokenizer
import pdfplumber

def extract_text_from_pdf(pdf_path):
    """Extract text from a PDF using pdfplumber"""
    extracted_text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            extracted_text += page.extract_text() + "\n"
    return extracted_text

def translate_text(text, model, tokenizer, device):
    """Translate text in chunks using MarianMTModel"""
    # Split the text into smaller chunks (e.g., paragraphs or sentences)
    chunk_size = 500  # Adjust the size based on the model's capacity
    chunks = [text[i:i+chunk_size] for i in range(0, len(text), chunk_size)]

    translated_text = ""

    for chunk in chunks:
        # Tokenize and translate the chunk
        inputs = tokenizer(chunk, return_tensors="pt", truncation=True, padding=True).to(device)
        with torch.no_grad():
            translated = model.generate(**inputs)
        translated_chunk = tokenizer.decode(translated[0], skip_special_tokens=True)
        translated_text += translated_chunk + "\n"

    return translated_text

def translate_pdf(pdf_path, source_language, target_language):
    """Translate the text from a PDF using MarianMTModel"""
    # Extract text from the PDF
    extracted_text = extract_text_from_pdf(pdf_path)

    # Load the model and tokenizer for the chosen language pair
    model_name = f'Helsinki-NLP/opus-mt-{source_language}-{target_language}'
    model = MarianMTModel.from_pretrained(model_name)
    tokenizer = MarianTokenizer.from_pretrained(model_name)

    # Check if GPU is available, else use CPU
    device = "cuda" if torch.cuda.is_available() else "cpu"
    model = model.to(device)

    # Translate the extracted text
    translated_text = translate_text(extracted_text, model, tokenizer, device)

    return translated_text

# Example usage
pdf_path = 'Shivam_Sharma_Programming_Resume.pdf'
translated_pdf_text = translate_pdf(pdf_path, 'en', 'de')  # English to German
print(translated_pdf_text)
```
